```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/omarmendoza564/datos/main/datos/titanic.csv')
df
```

```python
# 1. Si la edad es null (na) llenar el campo con la media de la edad (usar el metodo fillna())

df['age'].isnull().sum()
```

```python
df['age'] = df['age'].fillna(df['age'].mean()) 
```

```python
df['age'].isnull().sum()
```

```python
df['age']
```

```python
# 2. Si embarqued es na llenar con S


df['embarked'].isnull().sum()
```

```python
df['embarked'] = df['embarked'].fillna('S') 
```

```python
df['embarked'].isnull().sum()
```

```python
# 3. Cambiar los valores alfanumericos de sex a numero usando { 'female': 1, 'male': 0 }
df['sex'] = df['sex'].replace({'female':1, 'male':0})
```

```python
df['sex']
```

```python
# 4. Cambiar los valores alfanumericos de embarked a numero usando { 'S': 0, 'C': 1, 'Q': 2 }
df['embarked'] = df['embarked'].replace({'S':0, 'C':1, 'Q':2})
```

```python
df['embarked']
```

```python
# 5. Crear una nueva variable llamada child con valor 1 para los pasajeros con una edad menor a 16 años y 0 si el pasajero tiene 16 años o mas
df['child'] = df['age']
df
```

```python
df.loc[df['child'] < 16, 'child'] = 1
df.loc[df['child'] > 15, 'child'] = 0
```

```python
df.loc[[50, 40], ['name', 'age']]
```

```python
df.iloc[[50, 40]][['name', 'age']]
```

```python
# 6. Eliminar las columnas ticket, cabin y name
df.drop([ 'ticket','cabin', 'name'], axis=1, inplace=True)
```

```python
df
```

```python
# 7. Validar que no existan registros duplicados
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
df.duplicated().sum()
```

```python
# 8. Listar los valores unicos del campo sex
df['sex'].unique()
```

```python
# 9. Quitar los espacios vacios de los campos que no sean numericos
df['home.dest']= df['home.dest'].str.strip()
```

```python
df
```

```python
# 10.Guardar los datos limpios en un archivos json orientado a registros llamado iniciales_titanic_clean.json
df.to_json('datos/df/JCMT_titanic_clean.json',orient="records" 
           ,indent=2)
```
